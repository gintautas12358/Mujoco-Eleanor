# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
results = np.zeros((len(testing_set), len(plugs)))
labels = np.zeros(len(testing_set))
acc = 0
for i in range(len(testing_set)):
    inp_mat, label = testing_set[i]
    output, count = net(inp_mat.unsqueeze(0).to(device))
    if "ext" in dataset_path : #because network for extended results is shifted
        old_result = np.sum(output.squeeze(0).detach().numpy(), axis=1) 
        results[i] = (old_result+1) % 4
    else :
        results[i] = np.sum(output.squeeze(0).detach().numpy(), axis=1)
    labels[i] = label
    if np.argmax(results[i]) == labels[i] :
        acc+=1
print("Accuracy : ", acc/len(results))
```

```python
true_cl = labels
pred_cl = np.argmax(results, axis=1)
cm = confusion_matrix(true_cl, pred_cl, normalize="true")
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=plugs)
#disp.ax_ = ax
#disp.figure_ = fig
disp.plot(ax=ax)
ax.xaxis.set_tick_params(labelsize=16)
ax.yaxis.set_tick_params(labelsize=16)
plt.xlabel('Predicted Class', fontsize=16)
plt.ylabel('True Class', fontsize=16)
plt.title("Confusion Matrix Test", fontsize=20)
#plt.savefig('daily_action_pretrain_test_confusion_matrix_2000.png', bbox_inches='tight')
plt.show()
```

# Visualize the network output

Here, we will use `slayer.io.tensor_to_event` method to convert the torch output spike tensor into `slayer.io.Event` object and visualize a few input and output event pairs.

Load test data and run inference

```python
no_samples = 5
```

```python
for i in range(no_samples):
    output, count = net(inp[i].unsqueeze(0).to(device))
    inp_event = slayer.io.tensor_to_event(inp[i].cpu().data.numpy().reshape(2, 128, 128, -1))
    out_event = slayer.io.tensor_to_event(output.cpu().data.numpy().reshape(1, 4, -1))
    inp_anim = inp_event.anim(plt.figure(figsize=(5, 5)), frame_rate=24)
    out_anim = out_event.anim(plt.figure(figsize=(10, 5)), frame_rate=24)
    inp_anim.save(f'gifs/inp_{i}.gif', animation.PillowWriter(fps=24), dpi=300)
    out_anim.save(f'gifs/out_{i}.gif', animation.PillowWriter(fps=24), dpi=300)
    print(inp[i].unsqueeze(0).size())
    print(output.size())
    #for users with narrow GPU capacity
    #del output
    #torch.cuda.empty_cache()
```
